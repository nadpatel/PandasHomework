```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed


gender_count = purchase_df.groupby(['Gender'])[['Gender']].count()
gender_percentage = (purchase_df.groupby('Gender').size()/purchase_df['Gender'].count())*100
gender_count['Percentage of Players'] = gender_percentage 

gender_count.rename(index=str, columns={"Gender": "Total Count"})

gender_count
```

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

gender_purchase_count= purchase_df.groupby(['Gender'])[['Gender']].count()
gender_purchase_count.rename(index=str, columns={"Gender": "Purchase Count"})

avg_purchase_price = purchase_df.groupby(['Gender'])[['Price']].mean()
total_purchase_value = purchase_df.groupby(['Gender'])[['Price']].sum()
#avg_total_per_person = purchase_df.groupby('')

gender_purchase_count['Average Purchase Price'] = avg_purchase_price
gender_purchase_count['Total Purchase Value']= total_purchase_value
#gender_purchase_count['Avg Total Purchase Per Person']= avg_total_per_person



gender_purchase_count
```

```python
#Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]

#Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points
#Display Age Demographics Table

group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_df["Age Ranges"]= pd.cut(purchase_df["Age"], bins, labels = group_names)
purchase_df

age_counts = purchase_df.groupby(["Age Ranges"]).count()["Age"]
age_counts.rename(columns={"Age":"Total Count"})
age_counts = pd.DataFrame(age_counts)
lengthofbins = pd.value_counts(bins)
age_percentage= (age_counts/lengthofbins.count())
age_counts["Total Percentage"]= age_percentage

age_counts







```

```python
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

# Purchase Count	Average Purchase Price	Total Purchase Value	Avg Total Purchase per Person

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_df["Age Group"] = pd.cut(purchase_df["Age"], bins, labels=group_names)

purchase_age = purchase_df.groupby("Age Group")

#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
age_summary= pd.DataFrame(purchase_age["Purchase ID"].count())
age_summary = age_summary.rename(columns={"Purchase ID":"Purchase Count"})

avg_purch_price_age = round(purchase_age['Price'].sum()/purchase_age["Purchase ID"].nunique(),2)
age_summary['Average Purchase Price'] = avg_purch_price_age

total_purch_price_age = purchase_age['Price'].sum()
age_summary['Total Purchase Value'] = total_purch_price_age

purchase_per_person_age = round(purchase_age['Price'].sum()/purchase_age["SN"].nunique(),2)
age_summary['Avg Total Purchase per Person'] = purchase_per_person_age

#Optional: give the displayed data cleaner formatting
age_summary["Average Purchase Price"] = age_summary["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_summary["Total Purchase Value"] = age_summary["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_summary["Avg Total Purchase per Person"] = age_summary["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

#Display the summary data frame
age_summary



```

```python
#Run basic calculations to obtain the results in the table below
#Create a summary data frame to hold the results
#Sort the total purchase value column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

#Purchase Count	Average Purchase Price	Total Purchase Value

top_spend_subset = purchase_df.groupby(['SN'])
top_spender= pd.DataFrame(top_spend_subset["Purchase ID"].nunique())
top_spender = top_spender.rename(columns={"Purchase ID":"Purchase Count"})

avg_purch_top = round(top_spend_subset['Price'].sum()/top_spend_subset["Purchase ID"].nunique(),2)
top_spender['Average Purchase Price'] = avg_purch_top

tot_purch_price_age = top_spend_subset['Price'].sum()
top_spender['Total Purchase Value'] = tot_purch_price_age

#Optional: give the displayed data cleaner formatting
top_spender["Average Purchase Price"] = top_spender["Average Purchase Price"].astype(float).map("${:,.2f}".format)
top_spender["Total Purchase Value"] = top_spender["Total Purchase Value"].astype(float).map("${:,.2f}".format)

#Sort the total purchase value column in descending order
top_spender = top_spender.sort_values("Purchase Count", ascending=False)

#Display a preview of the summary data frame
top_spender.head(5)
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
#Create a summary data frame to hold the results
#Sort the purchase count column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

most_popular = purchase_df.groupby(['Item ID', 'Item Name'])

#Create a summary data frame to hold the results
most_popular_df = pd.DataFrame(most_popular["Purchase ID"].nunique())
most_popular_df = most_popular_df.rename(columns={"Purchase ID":"Purchase Count"})


item_price = most_popular['Price'].sum()/most_popular['Purchase ID'].nunique()
most_popular_df ['Item Price'] = item_price

total_value = most_popular['Price'].sum()
most_popular_df ['Total Purchase Value'] = total_value

#Sort the purchase count column in descending order

most_popular_df_sorted1 = most_popular_df.sort_values("Purchase Count", ascending=False)


#Optional: give the displayed data cleaner formatting
most_popular_df_sorted1["Item Price"] = most_popular_df_sorted1["Item Price"].astype(float).map("${:,.2f}".format)
most_popular_df_sorted1["Total Purchase Value"] = most_popular_df_sorted1["Total Purchase Value"].astype(float).map("${:,.2f}".format)


#Display a preview of the summary data frame
most_popular_df_sorted1.head(5)
```
